# Titanic

Dataset: https://www.kaggle.com/c/titanic/overview

#### Importando os Pacotes Necessários

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

#### Lendo os dataset

```python
df = pd.read_csv('./dataset/train.csv')
```

```python
df.head()
```

```python
#Excluindo algumas colunas desnecessárias
df = df.drop(['Name','Ticket','Cabin','Embarked'], axis=1)
```

```python
df.head()
```

#### Editando Chaves e Variáveis Resposta

```python
# Colocando uma Chave forte e renomeando a variável resposta como target
df = df.set_index(['PassengerId'])
df = df.rename(columns = {'Survived' : 'target'}, inplace = False)
```

```python
df.head()
```

#### Descritiva

```python
df.describe()
```

```python
#retorna quantos valores distintos temos em uma variável não numérica
df.describe(include=['O'])
```

#### Transformação dos dados

```python
#transforma os dados categóricos em 0 e 1 para ser aceito pelo algorítmo
df['Sex_F'] = np.where(df['Sex'] == 'female', 1, 0)

df['Pclass_1'] = np.where(df['Pclass'] == 1, 1, 0)
df['Pclass_2'] = np.where(df['Pclass'] == 2, 1, 0)
df['Pclass_3'] = np.where(df['Pclass'] == 3, 1, 0)
```

```python
df.head()
```

```python
df = df.drop(['Pclass', 'Sex'], axis = 1)
```

```python
df.head()
```

```python
#verificando a existência de núlos
df.isnull().sum()
```

```python
#preenchendo os valores nulos com 0
df.fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

#### Amostragem

```python
#separando 70% treino e 30% teste
x_train, x_test, y_train, y_test = train_test_split(df.drop(['target'], axis = 1),
                                                   df["target"],
                                                   test_size = 0.3,
                                                   random_state = 1234)
[{'treino': x_train.shape},{'teste': x_test.shape}]
```

#### Modelo (Random Forest)

```python
#1000 ávores desça 5 de profundidade
rndforest = RandomForestClassifier(n_estimators=1000,
                                  criterion='gini',
                                  max_depth= 5)
rndforest.fit(x_train,y_train)
```

```python
probabilidade = rndforest.predict_proba(df.drop('target', axis = 1))[:,1]
classificacao = rndforest.predict(df.drop('target', axis = 1))
```

```python
df['probabilidade'] = probabilidade
df['classificacao'] = classificacao
```
